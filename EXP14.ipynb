```python
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,4500],
    'Units_Sold':[25,60,35,80,50,15]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Data Running
bins1 = [0,20000,30000,60000]
labels1 = ['Low','Medium','High']
df['Price_Category'] = pd.cut(df['Price'],bins = bins1,labels =labels1)
print(df)
```

```python
#Data Running
bins2 = [0,30,60,100]
labels2 = ['Low Sales','Medium Sales','High Sales']
df['Sales_Category'] = pd.cut(df['Units_Sold'],bins = bins2,labels =labels2)
print(df)
```

```python
#Convert data type

print(df.dtypes)

#df['Units_Sold'] = df['Units_Sold'].astype(float)
```

```python
print(df.dtypes)
df['Units_Sold'] = df['Units_Sold'].astype(float)
print(df.dtypes)
df
```

```python
#Convert Text to Uppercase
df['Product'] = df['Product'].str.upper()
print(df)
```

```python
df['Price'] = df['Price'].round(2)
print(df)
```

```python
#Sort Data
sorted_df = df.sort_values(by='Price')
print(sorted_df)
```

```python
df.sort_values(by='Units_Sold', inplace=True)
print(df)
```

```python
df.sort_values(by='Units_Sold',ascending=False, inplace=True)
print(df)
```

```python
#Display Unique Values
print(df['Sales_Category'].unique())
```

```python
#Dataset: Food Delivery Orders
import pandas as pd
import numpy as np
data = {
'Order_ID': ['01', '02', '03', '04', '05', '06', '07', '08'],
'Order_Value': [250, 450, 120, 600, 300, 150, 500, 220],
'Delivery_Time': [30,45,25,60,35,20,50,28],
'Distance_km': [5,8,3,10,6,2,9,4]
}
df= pd.DataFrame(data)
df
```

```python
#Binning Order Value

bins = [0,200,400,700]

labels = ['Low Order', 'Medium Order', 'High Order']

df['Order_Category'] = pd.cut(df['Order_Value'], bins=bins, labels=labels)

print(df)
```

```python
#Binning Delivery Time

bins2 = [0,30,45,70]

labels2 = ["Fast", "Normal", "Slow"]

df['Delivery Speed'] = pd.cut(df['Delivery_Time'], bins=bins2, labels=labels2)

df
```

```python
# Binning Distance_km

bins_distance = [0, 4, 8, 12]

labels_distance = ["Short Distance", "Medium Distance", "Long Distance"]

df['Distance_Category'] = pd.cut(df['Distance_km'], bins=bins_distance, labels=labels_distance)

df
```

```python
# Count of categories

print(df['Order_Category'].value_counts())
df
```
